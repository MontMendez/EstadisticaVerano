# MARCO TEORICO

Regresión Lineal Simple: aplicado a un modelo describe la relación entre 2 variables cuantitativas, una dependiente y una independiente. La regresión permite estimar como cambia una variable dependiente a medida que cambia la variable independiente.

Según Herrera (2016), variable es una palabra que representa aquello que esta sujeto a algún tipo de cambio.

La variable independiente "x" es la causa, la que se manipula para ver lo que sucede en la variable dependiente "y", la cual, es el efecto, la que va cambiando dependiendo de "x".

Según Hal Varian (2006), el precio de las laptops puede variar dependiendo de las caracteristicas técnicas, la marca, la reputación y muchos otros factores que afectan en su precio de venta.



## Pruebas de hipotesis

ANOVA: Nos permite saber si hay relación entre variables o si hay diferencia significativa entre grupos.

     Ho: B1=0
     Hi: B1=/0
Donde B1=0 significa que no hay relación.

Shapiro: Nos permite saber si la distribución es normal en la recta.

     Ho: Los datos siguen una distribución normal
     Hi: Los datos no siguen una distribución normal

Breusch-Pagan: Utilizado para detectar si hay homoscedasticidad o heteroscedasticidad en regresion lineal.

     Ho: Hay homoscedasticidad
     Hi: Hay heteroscedasticidad

## Comprobación de supuestos

Test de shapiro: Comprobar si la distribución es normal.


Grafico  X  y  Y : Observar si los datos soportan la suposición de linealidad.


Gráfico de residuales: Observar si los datos soportan la suposición de linealidad y conocer el coeficiente de correlación.


Test de Breusch-Pagan: Detectar homoscedasticidad o heteroscedasticidad.

# Variables

## Company

```python
from statsmodels.formula.api import ols
import statsmodels.api as sm
modelo_lineal = ols('Price ~ C(Company)', data = df).fit()
tabla_anova = sm.stats.anova_lm(modelo_lineal)
tabla_anova
#se rechaza ho, hay diferencia al menos en un grupo, relación entre variables.
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt

nivel_de_significancia = 0.05

# Prueba de Tukey
tukey = pairwise_tukeyhsd(endog=df['Price'], groups=df['Company'], alpha=nivel_de_significancia)

# Mostrar los resultados
print(tukey)

# Gráfico de las diferencias entre grupos
tukey.plot_simultaneous(ylabel="Company", xlabel="Price")

plt.gca().spines['right'].set_visible(False) # derecha
plt.gca().spines['top'].set_visible(False)   # superior
plt.title("Múltiples comparaciones entre todos los pares (Tukey)")
plt.axvline(x=1050, color='red', linestyle='--')
plt.axvline(x=1900, color='green', linestyle='--')
```

#Regresión lineal simple

```python
x = df["RAM"]
y = df["Price"]
print("\n")

print("2 Realice un diagrama de dispersión para estos datos.")
import matplotlib.pyplot as plt
plt.figure(figsize = (8, 4), dpi = 100)
plt.scatter(x, y, label = "Price", color = 'pink')
plt.title("Diagrama de dispersión")
plt.xlabel('RAM') # nombrar eje x
plt.ylabel('Price') # nombrar eje y
plt.gca().spines[["top","right"]].set_visible(False)
print("\n")
plt.show()

print("3 ¿Los datos soportan la suposición de linealidad?")
print("No, no soportan la linealidad ya que no parecen seguir una secuencia")
print("\n")
plt.show()

print("4 Calcule el coeficiente de correlación e interprete el resultado.")
from scipy.stats import pearsonr
r, _ = pearsonr(x, y)
print(f'Coeficiente de correlación: {r: 0.4f}\n')
print("La relación es positiva, por lo tanto la tendencia se suele ver hacia arriba")
print("\n")
plt.show()

print("6 Obtenga la recta de regresión ajustada y grafíquelo sobre el gráfico de dispersión.")
import matplotlib.pyplot as plt
plt.figure(figsize=(8,4),dpi=100)
plt.scatter(x,y,
            label="Price",
            color="pink")
plt.title("Recta de regresión ajustada sobre gráfico de dispersión")
plt.xlabel("RAM")
plt.ylabel("Price")
plt.legend()
plt.grid()
plt.gca().spines[["top","right"]].set_visible(False)
import statsmodels.api as sm
x_constante = sm.add_constant(x)
modelo = sm.OLS(y, x_constante).fit()
yc = modelo.predict(x_constante)
plt.plot(x, yc, color = 'grey', linestyle = '--')
print("\n")
plt.show()

print("5 Calcule el coeficiente de determinación e interprete el resultado.")
from sklearn.metrics import r2_score
r2 = r2_score(y, yc)
print(f'Coeficiente de determinación: {r2: 0.4f}\n')
print("La variación en Y es explicada en un 54.80% por la variable X")
print("\n")
plt.show()

print("7 Obtenga un intervalo de confianza del 95% para la pendiente de la recta de regresión ajustada ( b1 )")
nivel_de_confianza = 0.95
intervalo_de_confianza = modelo.conf_int(alpha = 1 - nivel_de_confianza)
intervalo_de_confianza_b1 = intervalo_de_confianza.iloc[1]
print(f'Intervalo de confianza para b1 de {nivel_de_confianza: 0.0%}')
print(f'{intervalo_de_confianza_b1[0]: 0.4f} < b1 < {intervalo_de_confianza_b1[1]: 0.4f}\n')
plt.show()

print("8 Calcule los residuales y trace un nuevo gráfico de dispersión. Comente, ¿Parece que se verifican los supuestos?")
residuales = modelo.resid
plt.figure()
plt.scatter(x, residuales, color = 'pink')
plt.title("Gráfico de dispersión de residuales")
plt.xlabel('RAM')
plt.ylabel('Residuales')
plt.axhline(y = 0, color = 'gray', linestyle = '--')
plt.gca().spines[['top', 'right']].set_visible(False)
print("........")
print("\n")
plt.show()

print("9 Realice la prueba de Shapiro para los residuales y comente el resultado.")
from scipy.stats import shapiro
_, valor_p_sh = shapiro(residuales)
print(f'valor-p de Shapiro: {valor_p_sh: 0.4f}\n')
print("Se rechaza ho, por lo tanto los datos no siguen una distribucción normal")
print("\n")
plt.show()

print("10 Realice la prueba de Brausch-Pagan para los residuales y comente el resultado.")
from statsmodels.stats.api import het_breuschpagan
_, valor_p_bp, _, _ = het_breuschpagan(residuales, x_constante)
print(f'valor_p de Breusch-Pagan: {valor_p_bp: 0.4f}\n')
print("Se rechaza ho, no hay homoscedasticidad, hay heteroscedasticidad")
print("\n")
plt.show()

print("11 Utiliza la recta de regresión para interpolar dos valores y extrapolar uno.")
print("interpolar:")
print(f'para x = 12, y = {modelo.predict([1, 12])}')
print(f'para x = 24, y = {modelo.predict([1, 24])}')
print("extrapolar:")
print(f'para x = 128, y = {modelo.predict([1, 128])}')
print("\n")
plt.show()

print("12 Realice una tabla ANOVA e interprete el resultado.")
from statsmodels.formula.api import ols
# Y ~ X
modelo_lineal = ols("Price ~ RAM", data = df).fit()
tabla_anova = sm.stats.anova_lm(modelo_lineal)
print(tabla_anova)
print("se rechaza ho, hay diferencia al menos en un grupo, relación entre variables.")
```

# CONCLUSIÓN

La matriz de correlación nos mostró que hay relación entre las variables. Utilizando el mapa de calor se observó una correlación positiva y bastante notable entre "RAM" y "Price". Con un valor de 0.74.
El coeficiente de correlación con un valor de 0.7403, nos indica que hay una relación algo fuerte entre las variables dependiente e independiente. Mientras que el coeficiente de determinación con un valor de 0.5480 nos dice que el 54.80% de la variación en "y" (Price), es explicada por la variable "x" (RAM).
Gracias al intervalo de confianza observamos que por cada aumento de la memoria RAM, el precio sube aproximadamente entre 96.68 y 106.84 euros, con un nivel de significancia de 5% y de confianza 95%. Claro que no podemos estar muy seguros debido a que los supuestos de normalidad (Shapiro) y de homoscedasticidad (BreuschPagan) no se cumplen.

Finalmente, puedo decir que este modelo de regresión lineal simple no es el más adecuado para detectar o pronosticar los posibles precios de las laptops, es necesario un modelo más complejo y servible para que haya mejor relación entre las variables y por lo tanto una mejor aplicación.

Por otra parte,  ANOVA confirma la relación significativa entre "Price" y "RAM"; sin embargo, como he dicho anteriormente, no se cumplen los supuestos de normalidad y homoscedasticidad, por lo que se debe tener precaución al utilizar este modelo e interpretar los resultados dados por las variables, esto indica que se debe usar un modelo más complejo y avanzado para una predicción en el precio de las laptops más claro y preciso.
